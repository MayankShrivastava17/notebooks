```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.countplot(data=data,x=data["diagnosis"])
plt.title('Different type of diagnosis')
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()

data['diagnosis'] = le.fit_transform(data['diagnosis'])

data['diagnosis'].head()
```

```python
data.head()
```

```python
data.tail()
```

```python
x = data.drop(columns="diagnosis",axis=1).values
y = data["diagnosis"].values
```

```python
x.shape
```

```python
y.shape
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, shuffle=True)
```

```python
print(x_train.shape)
print(y_train.shape)
print(x_test.shape)
print(y_test.shape)
```

<h1 style="text-align:center;font-size:25px;font-weight:bolder;">K-Nearest Neighbour</h1>

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
from sklearn.metrics import accuracy_score

for i in range(1, 20):
    knn = KNeighborsClassifier(n_neighbors=i)
    knn.fit(x_train, y_train)
    pred = knn.predict(x_test)
    acc = accuracy_score(pred, y_test)
    print ('n_neighbour : {} -> Prediction : {}'.format(i, acc))
```

```python
final = KNeighborsClassifier(n_neighbors=2)
final.fit(x_train, y_train)
```

```python
pred = final.predict(x_test)
accuracy_score(pred, y_test)
```

<h1 style="text-align:center;font-size:25px;font-weight:bolder;">Neural Network</h1>

```python
import tensorflow as tf
import keras
from keras.layers import Dense
from keras.models import Sequential 
```

```python
model = Sequential([
    Dense(64, activation='relu', input_shape=(30,), name="input-layer"),
    Dense(32, activation='relu', name="hidden-layer-1"),
    Dense(16, activation='relu', name="hidden-layer-2"),
    Dense(8, activation='relu', name="hidden-layer-3"),
    Dense(1, activation='sigmoid')
])
```

```python
model.compile(
    loss='binary_crossentropy',
    optimizer='adam', 
    metrics=['accuracy']
)
```

```python
model.summary()
```

```python
from keras.utils.vis_utils import plot_model
plot_model(model, to_file='model_plot.png', show_shapes=True, show_layer_names=True)
```

```python
model.fit(
    x_train, 
    y_train,
    batch_size=10,
    epochs=10
)
```

```python
model.evaluate(x_test, y_test)
```

```python
pred = model.predict(x_test)
```

```python
plt.plot(pred, '--r')
plt.plot(y_test)
plt.title('Neural Network')
plt.show()
```
