```python
# import pandas to read the dataset 

import pandas as pd
```

```python
# import the dataset 

data = pd.read_csv('fish.csv')
```

```python
# convert the dataset into Data Frame 

df = pd.DataFrame(data)
```

```python
# to visualize the dataset 

import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# the species of fish in the dataset 

sns.set_theme()
sns.countplot(data=data, x=data["Species"])
plt.title('Species of Fishes in the dataset')
plt.show()
```

```python
# complete dataset visualization 

sns.pairplot(df, hue='Species')
```

```python
# encode the species into numerical values to fit in the model 

from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

data['Species'] = le.fit_transform(data['Species'])

data
```

```python
# split the target and dependent values 

y = data['Species'].values
x = data.drop(columns='Species', axis=1).values
```

```python
x
```

```python
y
```

```python
# import train_test_split to split the dataset into training and testing 

from sklearn.model_selection import train_test_split
```

```python
# split the dataset into train and test dataset 

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=27, shuffle=True)
```

```python
# checking the shape of training and testing dataset 

x_train.shape, y_train.shape, x_test.shape, y_test.shape
```

```python
# import the logistic regression model 

from sklearn.linear_model import LogisticRegression
```

```python
# define the logistic regression model 

reg = LogisticRegression(solver='liblinear', random_state=0)
```

```python
# fit and train the logistic regression model 

reg.fit(x_train, y_train)
```

```python
# plotting the intercept of the model 

sns.set_theme()
plt.plot(reg.intercept_, 'b')
plt.title('Model Intercept')
plt.show()
```

```python
# plotting the coefficient of the model 

sns.set_theme()
plt.plot(reg.coef_, 'r')
plt.title('Model Coefficient')
plt.show()
```

```python
# checking the model accuracy score 

from sklearn.metrics import accuracy_score

pred = reg.predict(x_test)

print(f'Model Prediction is { round(accuracy_score(y_test, pred) * 100, 2) }%')
```

```python
# training score and testing score 

print(f'Model Training Score { round(reg.score(x_train, y_train) * 100, 2) }%')
print(f'Model Test Score { round(reg.score(x_test, y_test) * 100 , 2) }%')
```
